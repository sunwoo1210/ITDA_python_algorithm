#### 복합문의 구조
- 헤더: 복합문의 첫 부분(if/while로 시작하여 :으로 끝나는 부분)
- 스위트: 헤더와 한 세트로 따라다니는 실행문 <br><br>

* if식: 스위트 (if문 - 반드시 1개 필요)
* elif식: 스위트 (elif문 - 없거나, 있으면 n개 가능)
* else식: 스위트 (else문 - 없거나, 있으면 1개 가능)

```python
#실습 1-2
#세 정수의 최댓값 구하기

def max3(a, b, c):
    maximum = a
    if b > maximum:
        maximum = b
    if c > maximum: 
        maximum = c
    return maximum  #최댓값 반환

print(f'max3(3, 2, 1) = {max3(3, 2, 1)}') #a>b>c
print(f'max3(3, 2, 2) = {max3(3, 2, 2)}') #a>b=c
print(f'max3(3, 1, 2) = {max3(3, 1, 2)}') #a>c>b
print(f'max3(3, 2, 3) = {max3(3, 2, 3)}') #a=c>b
print(f'max3(2, 1, 3) = {max3(2, 1, 3)}') #c>a>b
print(f'max3(3, 3, 2) = {max3(3, 3, 2)}') #a=b>c
print(f'max3(3, 3, 3) = {max3(3, 3, 3)}') #a=b=c
print(f'max3(2, 2, 3) = {max3(2, 2, 3)}') #c>a=b
print(f'max3(2, 3, 1) = {max3(2, 3, 1)}') #b>a>c
print(f'max3(2, 3, 2) = {max3(2, 3, 2)}') #b>a=c
print(f'max3(1, 3, 2) = {max3(1, 3, 2)}') #b>c>a
print(f'max3(2, 3, 3) = {max3(2, 3, 3)}') #b=c>a
print(f'max3(1, 2, 3) = {max3(1, 2, 3)}') #c>b>a
```

#### 보충수업 1-5
#### 세 정수의 대소 관계와 중앙값
- 결정 트리: 예/아니오로 갈라 세부적으로 나눔, 조합을 나열한 모습이 나무 같음

```python
#실습 1c-2
#세 수를 입력받아 중앙값 구하기1

def med3(a, b, c):
    if a >= b:
        if b >= c:
            return b
        elif a <= c:
            return a
        else:
            return c
    elif a > c:
        return a
    elif b > c:
        return c
    else:
        return b
    
print('세 정수의 중앙값을 구합니다.')
a = int(input('정수 a의 값을 입력하세요.: '))
b = int(input('정수 b의 값을 입력하세요.: '))
c = int(input('정수 c의 값을 입력하세요.: '))

print(f'중앙값은 {med3(a, b, c)}입니다.')
```

```python
#실습 1c-2 
#세 수를 입력받아 중앙값 구하기2
def med3(a, b, c):
    if (b >= a and c <= a) or (b <= a and c >= a):
        return a
    elif (a > b and c < b) or (a < b and c > b):  #위에 if문 조건을 뒤집은 판단을 elif에서 수행 -> 비효율적
        return b
    return c

print('세 정수의 중앙값을 구합니다.')
a = int(input('정수 a의 값을 입력하세요.: '))
b = int(input('정수 b의 값을 입력하세요.: '))
c = int(input('정수 c의 값을 입력하세요.: '))

print(f'중앙값은 {med3(a, b, c)}입니다.')
```

#### 연산자와 피연산자
- 산술 연산자: +, - 등의 기호
- 피연산자: 연산 대상 <br><br>

- 단항 연산자: 피연산자 1개 (ex. -a)
- 이항 연산자: 피연산자 2개 (ex. a<b)
- 삼항 연산자: 피연산자 3개 (ex. 조건 연산자, a if b else c)

#### 순서도 기호
<프로그램 순서도>
- 실제로 실행할 연산을 나타내는 기호
- 제어 흐름을 나타내는 선 기호
- 프로그램 순서도를 이해하거나 작성하는 데 편리한 특수 기호 

- 데이터 (평행사변형)
: 기억 장치를 지정하지 않은 데이터 자체 <br><br>

- 처리 (직사각형)
: 여러 종류의 처리 기능 (연산, 집합의 실행) <br><br>

- 미리 정의한 처리 (직사각형 양 옆에 선)
: 다른 곳에서 이미 정의한 하나 이상의 연산 또는 명령으로 이루어진 처리 <br><br>

- 판단 (마름모)
: 하나의 입구와 하나 이상을 선택하는 출구 있음, 판단 기호 안에 정의한 조건을 평가하여 하나의 출구를 선택하는 판단 기능. 주로 예상되는 평가 결과는 경로를 나타낸 선 가까이에 표기 <br><br>

- 루프 범위 (각진 반원형 마주보고)
: 루프의 시작 기호와 종료 기호로 이루어짐, 루프 범위 기호 안에는 같은 이름(내용)을 표기 <br><br>

- 선 (선, 화살표)
: 제어의 흐름을 의미, 흐름의 방향을 분명히 나타낼 때는 화살표 사용 <br><br>

- 단말 (둥근 직사각형)
: 외부 환경으로 나가거나 외부 환경에서 들어오는 것을 나타냄, 주로 프로그램 흐름의 시작과 종료 나타냄
